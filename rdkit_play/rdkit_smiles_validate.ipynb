```python
from rdkit.Chem import AllChem
from rdkit import Chem, RDLogger
RDLogger.logger().setLevel(RDLogger.CRITICAL)
```

## Validate SMILES

```python
smiles = "Cc1ccnc1C"
```

```python
def is_valid_mol_to_smiles(smi):
    # Use MolFromSmiles without sanitize
    mol = Chem.MolFromSmiles(smi, sanitize=False)
    try:
        # Use SanitizeMol to do the sanitizion instead of MolFromSmiles
        _ = Chem.SanitizeMol(mol)
    except ValueError:
        None
        
    # All the output SMILES would be kekulized SMILES
    # so for the can not kekulized molecules will be changed to be unaromatic
    
    return Chem.MolToSmiles(mol, kekuleSmiles=True)
```

```python
valid_smiles = is_valid_mol_to_smiles(smiles)
valid_smiles
```

## Read Molecules

```python
valid_mol = Chem.MolFromSmiles(valid_smiles)
valid_mol
```

## Export MolFile

```python
mol_block = Chem.MolToMolBlock(valid_mol)
mol_block
```
