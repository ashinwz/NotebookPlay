```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit.Chem.EnumerateStereoisomers import EnumerateStereoisomers
m = Chem.MolFromSmiles('BrC=CC1OC(C2)(F)C2(Cl)C1')
isomers = tuple(EnumerateStereoisomers(m))
for isomer in isomers: 
    Chem.AssignStereochemistry(isomer,force=True,cleanIt=True)
```

```python
isomers[2]
```

```python
fp1=AllChem.GetMorganFingerprintAsBitVect(isomers[1], 4, 2048, useChirality=True)
fp2=AllChem.GetMorganFingerprintAsBitVect(isomers[2], 4, 2048, useChirality=True)
```

```python
fp1_o=AllChem.GetMorganFingerprintAsBitVect(isomers[1], 4, 2048)
fp2_o=AllChem.GetMorganFingerprintAsBitVect(isomers[2], 4, 2048)
```

```python
print(DataStructs.FingerprintSimilarity(fp1_o,fp2_o))
```

```python
fp1=AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(Chem.MolToSmiles(isomers[1])), 4, 2048, useChirality=True)
fp2=AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(Chem.MolToSmiles(isomers[2])), 4, 2048, useChirality=True)
print(DataStructs.FingerprintSimilarity(fp1,fp2))
```
